```python
avg_weight = round(data["weight"].mean(), 2)
avg_weight
```

```python
data["weight"] = data["weight"].fillna(avg_weight)
data
```

```python
data["birth_place"] = data["birth_place"].str.replace('NY', 'New York')
data.head(5)
```

```python
data["birth_place"] = data["birth_place"].str.replace('LA', 'Los Angeles')
data.head()
```

```python
data["weight"] = data["weight"] * 2.205
data.head()
```

#####Changing Weight Kg to Lb 

>>>Simple featured scaling

```python
data["height"] = data["height"]/data["height"].max()
data.head()
```

>>Normalized value= value/max(value)

Min-Max Normalization=(value-value.min())/(value.max()-value.min())

```python
data["weight"] = (data["weight"] - data["weight"].min()) / (data["weight"].max() - data["weight"].min())
data.head()
```

```python
data["z_scored_salary"] = (data["salary"] - data["salary"].mean()) /data["salary"].std()
data.head()
```

>>Grouping Salary Groups

```python
import numpy as np
bins = np.linspace(min(data["salary"]), max(data["salary"]), 6)
bins
```

```python
bin_names = ["Very Low", "Low", "Medium", "High", "Very High"]
data["salary-group"] = pd.cut(data["salary"], bins, labels=bin_names, include_lowest=True)
```
